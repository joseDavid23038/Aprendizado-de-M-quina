## Exploração dos datasets:

Inicialmente, a biblioteca a ser utilizada para carregar os datasets é importada. Para além dela, as biblitecas Pandas e Statistics foram importadas para o suprimento de demandas que pensei que precisariam delas para serem supridas, mas que notei ao longo da realização da atividade que todas poderiam ser resolvidas com a Seaborn.

```python
import seaborn as sbn
import statistics as st
import pandas as pd
```

A função get_dataset names é utilizada para exibir as etiquetas dos repositórios, as quais serão necessárias para carregá-los através da função load_data_set. <sup>[1]<sup>

```python
print(sbn.get_dataset_names())
```

```python
iris_plantas = sbn.load_dataset('iris', cache=True, data_home=None)
pinguins = sbn.load_dataset('penguins', cache=True, data_home=None)
gorjetas = sbn.load_dataset('tips', cache=True, data_home=None)
```

O primeiro datasets carrega é de nome "iris", o qual é referente a um conjunto de dados contém 3 classes de 50 instâncias cada, onde cada classe se refere a um tipo de planta íris. <sup>[5]</sup> Realizando a plotagem do dataframe por meio da função display<sup>[2]</sup>, é notável que esse dataset exibe informações de comprimento e largura das petalas e sepalas de diferentes espécies de iris. Por meio do dtypes, foi possível notar que as variáveis as quatro primeiras colunas se tratam de floats e a última de uma variável objeto, a qual engloba o string da espécie. Realizando a conversão dos tipos por meio da função convert_dtypes, as variáveis descritas como float continuaram sendo floats e a objeto se tornom string.



```python
display(iris_plantas)
print("\n Tipos de variável:\n", iris_plantas.dtypes)
print("\n Tipos de variável(convertidas):\n", iris_plantas.convert_dtypes().dtypes)
```

Realizando o plotagem de um histograma<sup>[7]</sup> da quantidade de plantas com um um determinado comprimento de petala é perceptível que a setosa tende a ter o menor comprimento de petala, enquanto a versicolor tende a possuir um tamanho médio em relação a setosa e a virginica, a qual apresenta os maiores tamanhos. Essa distribuição de tamanhos que se diferem pela espécie pode ser facilmente notada na imagem 1 logo abaixo do histograma.   

```python
ax = sbn.histplot(iris_plantas, x="petal_width", hue='species')
sbn.move_legend(ax, "center right")
```

O segundo dataset selecionado é referente a dados tamanhos de bico e asa(nanadeira) de três diferentes espécies de pinguim, a Gentoo, a Chinstrap e a Adelie, as quais habitam algumas ilhas da Península Antártica.<sup>[10]</sup>

Realizando a exbição do data frame e fazendo usando da função dtypes, é possível observar os diferentes tipos de variáveis de cada coluna. As colunass referentes a espécie, ilha e sexo são do tipo objeto, indicando que não são referente a variáveis quantitativas. As colunas referente a tamanho e altura do bico, massa corporal em gramas e tamanho da nadadeira seriam float. Ao realizar a correção dos tipos de variável, as que objeto se tornaram string e as das colunas de tamanho de nadadeira e massa corporal em gramas se tornaram valores do tipo inteiro. 

```python
display(pinguins)
print("\n Tipos de variável:")
display(pinguins.dtypes)
print("\n Tipos de variável(corrigidas):")
display(pinguins.convert_dtypes().dtypes)
```

Plotando um histograma de tamanho de bico por quantiades de pinguins de diferentes espécies em diferentes ilhas, é perceptível que a espécie Adelie tende a possuir o menor tamanho de bico e habita as três ilhas das quais as amotras foram tiradas e convivem com as duas outras espécies, enquanto as espécies Gentoo e Chinstrap apenas habitam uma ilha e não convivem entre elas. Outra coisa notável é que a Gentoo tende ter o maior tamanho de bico.

```python
g = sbn.displot(
    pinguins,
    x="bill_length_mm", hue="species",
    col="island", col_wrap=2, height=3,
)
sbn.move_legend(g, "upper left", bbox_to_anchor=(.55, .45))
```

Realizando a plotagem de uma gráfico de pontos do comprimento do bico por comprimento da nadadeira das diversas espécies, é possível visualizar o espalhamento das informações dos pinguins de diversas espécies. 

```python
sbn.relplot(pinguins, x="flipper_length_mm", y="bill_length_mm", hue='species')
```

Para quantificar a relação do tamanho do bico pelo tamanho da asa, é necessário fazer diferentes filtragens para analisar uma espécie por vez e também para conseguir remover as linhas em que não há informações referentes a tamanho de bico e nadadeira. O uso da função "dropna" do pandas foi feito para conseguir remover essas linhas sem valores, no entanto, a importação da biblioteca pandas acabaria não sendo necessária para fazer isso tendo em vista que a Seaborn também atua com esta função. <sup>[8][10]</sup> 
A função value_counts foi utilizada para contar a quantidades de pinguins de cada espécie que teriam seus dados quantitativos usados para calcular a correlação.<sup>[3]</sup> O cálculo da correlação é feito por meio da função corr, também da biblioteca Pandas, mas acessível pela Seaborn.<sup>[9]</sup> </br>
**Observação:** A remoção do Not a Number apenas seria necessária caso eu tivesse continuado com a biblitoeca statistics, pois a função "correlation" que executa esse cálculo nessa biblioteca não aceita o NaN. Considerando que fiz uso da função "corr" da biblioteca Seaborn, eu poderia apenas ter mantido o NaN e a função seria executada sem problema, e até cheguei a fazer isso na última linha do programa usando o dataframe do que apenas passou pela filtragem de espécie de pinguim.

```python
filtro = pinguins['species'] == "Gentoo"
pinguins_filtro = pinguins.loc[filtro]
correcao_asa = pinguins_filtro.dropna(subset=['flipper_length_mm'])
correcao_bico = correcao_asa.dropna(subset = ['bill_length_mm'])

filtro2 = pinguins['species'] == "Chinstrap"
pinguins_filtro2 = pinguins.loc[filtro2]
correcao_asa2 = pinguins_filtro2.dropna(subset=['flipper_length_mm'])
correcao_bico2 = correcao_asa2.dropna(subset = ['bill_length_mm'])

filtro3 = pinguins['species'] == "Adelie"
pinguins_filtro3 = pinguins.loc[filtro3]
correcao_asa3 = pinguins_filtro3.dropna(subset=['flipper_length_mm'])
correcao_bico3 = correcao_asa3.dropna(subset = ['bill_length_mm'])


#display(correcao_bico)
#display(correcao_bico2)
#display(correcao_bico3)

print(correcao_bico['species'].value_counts())
print(correcao_bico2['species'].value_counts())
print(correcao_bico3['species'].value_counts(),"\n")

correlacao = correcao_bico['flipper_length_mm'].corr(correcao_bico['bill_length_mm'])
print("A correlação entre tamanho da asa e do bico do pinguim Gentoo é de", correlacao)

correlacao2 = correcao_bico2['flipper_length_mm'].corr(correcao_bico2['bill_length_mm'])
print("A correlação entre tamanho da asa e do bico do pinguim Chinstrap é de", correlacao2)

correlacao3 = pinguins_filtro3['flipper_length_mm'].corr(pinguins_filtro3['bill_length_mm'])
print("A correlação entre tamanho da asa e do bico do pinguim Adelie é de", correlacao3)
```

Observando os resultados, é notável que a espécie Gentoo possui uma maior correlação do tamanho do bico com a asa, mostrando que, ao menos para esta amostra, essa espécie tende a exibir uma relação proporcional anatômica mais frequentemente.  

Relizando novamente uma plotagem multipla de histograms, mas dessa vez considerando o número de pinguins com diferentes massas corporais, machos e femeas de diferentes espécies, é possível notar que nas três espécies de pinguim, os machos tendem a possuir mais massa corporal que as fêmeas.

```python
g = sbn.displot(
    pinguins,
    x="body_mass_g", hue="sex",
    col="species", col_wrap=2, height=3,
)
sbn.move_legend(g, "upper left", bbox_to_anchor=(.55, .45))
```
